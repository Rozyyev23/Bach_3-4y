```python
import os

folder_path = "/content/yolo_labels/labels/train"

for file_name in os.listdir(folder_path):
    if file_name.endswith(".txt"):
        file_path = os.path.join(folder_path, file_name)
        new_contents = []

        with open(file_path, "r") as file:
            for line in file:
                if not line.startswith(("1", "2")): # меняем
                    new_contents.append(line)

        with open(file_path, "w") as file:
            file.writelines(new_contents)


folder_path_1 = "/content/yolo_labels/labels/val"

for file_name in os.listdir(folder_path_1):
    if file_name.endswith(".txt"):
        file_path = os.path.join(folder_path_1, file_name)
        new_contents = []

        with open(file_path, "r") as file:
            for line in file:
                if not line.startswith(("1", "2")): # меняем
                    new_contents.append(line)

        with open(file_path, "w") as file:
            file.writelines(new_contents)


folder_path_2 = "/content/yolo_labels/labels/test"

for file_name in os.listdir(folder_path_2):
    if file_name.endswith(".txt"):
        file_path = os.path.join(folder_path_2, file_name)
        new_contents = []

        with open(file_path, "r") as file:
            for line in file:
                if not line.startswith(("1", "2")): # меняем
                    new_contents.append(line)

        with open(file_path, "w") as file:
            file.writelines(new_contents)
```
