2 вопрос
Задан двухмерный массив ar1 размерности (25,4), состоящий из случайных целых чисел в пределах от 0 до 15. Определить, в каких столбцах не менее 5 раз встречается значение, максимальное по своей строке(вывести индексы этих столбцов на экран с соответствующим комментарием)
Решить задачу средствами numpy и/или pandas. Не использовать циклы и конструкции стандартного Python там, где можно использовать возможности данных библиотек.

```python
import pandas
import numpy as np
```

```python
ar1 = np.random.randint(low=0, high=16, size=(25, 4))
ar1
```

```python
max_in_rows = ar1.max(axis=1)
max_in_rows
```

```python
count_max_in_cols = (ar1 == max_in_rows.reshape(-1, 1)).sum(axis=0)
count_max_in_cols
```

```python
print(f'Индексы столбцов, которые не менее 5 раз встречается значение максимальное по своей строке {list(np.where(count_max_in_cols >= 5)[0])}')
```

3 Вопрос Необходимо найти все целые числа от 1 до 400_000, для которых sin от этого числа будет иметь не менее 9 одинаковых цифр(например sin(139198)=0.30763333543133703 т.е. имеет 9 троек). Ответ представляет собой список пар(целое число, значение синуса для него), для которых значение синуса удовлетворяет условию.
Решение этой задачи распараллелить, используя multiprocessing Pool. Сравнить продолжительность последовательного и параллельного решения задачи

```python
%%time
#Последовательное решение задачи
import math

result = []

for num in range(1, 400_001):
    sine = str(math.sin(num))
    for char in sine:
        if sine.count(char) >= 9:
            result.append((num, float(sine)))
            break
            
print(result)
```

```python
num_lists = [[i for i in range(1, 100_001)],[i for i in range(100_001, 200_001)],[i for i in range(200_001, 300_001)],[i for i in range(300_001, 400_001)]]
len(num_lists[0]),len(num_lists[1]),len(num_lists[2]),len(num_lists[3])
```

```python
%%file func1.py

import math

def func(num_list):
    result = []
    for number in num_list:
        sine = str(math.sin(number))
        for char in sine:
            if sine.count(char) >= 9:
                result.append((number, float(sine)))
                break
    return result
```

```python
from func1 import func
```

```python
%%time

import multiprocessing

with multiprocessing.Pool(processes = 4) as pool:
    res = pool.map(func, num_lists)
res
```
