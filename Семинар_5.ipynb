### Задача линейного программирования


Построить рандомный граф. Выбрать вершины, которые будут являться началом и концом и найти кратчайший путь между ними.


```python
import random
import networkx as nx
import matplotlib.pyplot as plt
```

```python
class Graph:
    def __init__(self, num_vertices):
        self.num_vertices = num_vertices
        self.graph = nx.Graph()

    def add_edge(self, u, v, weight):
        self.graph.add_edge(u, v, weight=weight)

    def generate_random_graph(self, num_edges):
        if num_edges > self.num_vertices * (self.num_vertices - 1) // 2:
            sys.exit()

        edges_added = 0
        while edges_added < num_edges:
            u = random.randint(0, self.num_vertices - 1)
            v = random.randint(0, self.num_vertices - 1)
            if u != v and not self.graph.has_edge(u, v):
                weight = random.randint(1, 10)
                self.add_edge(u, v, weight)
                edges_added += 1

    def dijkstra(self, start_vertex, end_vertex):
        shortest_path = nx.shortest_path(self.graph, start_vertex, end_vertex, weight='weight')
        return shortest_path

    def visualize_graph(self, shortest_path=None):
        pos = nx.spring_layout(self.graph)
        nx.draw_networkx_nodes(self.graph, pos, node_size=500, node_color='lightblue')
        nx.draw_networkx_edges(self.graph, pos)
        nx.draw_networkx_labels(self.graph, pos, font_size=10, font_color='black')
        if shortest_path:
            edges = [(shortest_path[i], shortest_path[i+1]) for i in range(len(shortest_path)-1)]
            nx.draw_networkx_edges(self.graph, pos, edgelist=edges, edge_color='red', width=2.0)
        plt.show()
```

```python
# Пример использования
num_vertices = int(input('Введите кол-во вершин в графе:'))
num_edges = int(input('Введите кол-во ребер в графе:'))

g = Graph(num_vertices)
g.generate_random_graph(num_edges)

print("Сгенерированный граф:")
g.visualize_graph()

start_vertex = int(input('Введите начальную вершину:'))
end_vertex = int(input('Введите конечную вершину:'))

print("Начальная вершина:", start_vertex)
print("Конечная вершина:", end_vertex)

shortest_path = g.dijkstra(start_vertex, end_vertex)
print("Кратчайший путь:", shortest_path)

g.visualize_graph(shortest_path)
```
