```python
import yfinance as yf
import pandas as pd
import statsmodels.api as sm

# Получение данных о ценах акций
data = yf.download(['AMZN', 'TSLA', 'NVDA', 'JPM', 'JNJ'], start='2018-01-01', end='2022-12-31')['Close']

# Создание зависимой переменной (y) и независимых переменных (X)
y = data['AMZN']
X = data[['TSLA', 'NVDA']]

# Добавление константы к матрице независимых переменных
X = sm.add_constant(X)

# Создание и обучение модели линейной регрессии
model = sm.OLS(y, X)
results = model.fit()

# Просмотр результатов оценки модели
print(results.summary())
```

Структурная форма модели:

```python
# Получение данных о ценах акций
data = yf.download(['AMZN', 'TSLA'], start='2022-01-01', end='2022-12-31')['Close']

# Создание зависимой переменной (y) и независимой переменной (X)
y = data['AMZN']
X = data['TSLA']

# Добавление константы к матрице независимых переменных
X = sm.add_constant(X)

# Создание и обучение модели структурной формы
model = sm.OLS(y, X)
results = model.fit()

# Просмотр результатов оценки модели
print(results.summary())
```

Структурная форма модели показывает следующие результаты:


R-квадрат является мерой соответствия модели данным. Значение 0.736 означает, что около 73.6% вариации цены акции Apple может быть объяснено изменениями цены акции Microsoft.
Durbin-Watson статистика: 0.087

Durbin-Watson статистика используется для проверки автокорреляции остатков модели. Значение 0.087 указывает на сильную положительную автокорреляцию.
Omnibus, Jarque-Bera и Kurtosis тесты:

Omnibus и Jarque-Bera тесты используются для проверки нормальности остатков модели. Здесь Omnibus тест и Jarque-Bera тест показывают низкие p-значения, что говорит о нарушении предпосылки о нормальности остатков. Kurtosis тест показывает, что остатки имеют отклонение от нормального распределения (значение куртозиса 2.388).

### Линейная регрессионная модель:


```python
# Создание и обучение линейной регрессионной модели
model_linear = sm.OLS(y, X)
results_linear = model_linear.fit()

# Просмотр результатов оценки линейной регрессионной модели
print(results_linear.summary())
```

### Нелинейная регрессионная модель:


```python
# Создание и обучение нелинейной регрессионной модели
model_nonlinear = sm.OLS(y, sm.add_constant(X**2))
results_nonlinear = model_nonlinear.fit()

# Просмотр результатов оценки нелинейной регрессионной модели
print(results_nonlinear.summary())
```

### Линейная модель парной регрессии:


```python
# Создание и обучение линейной модели парной регрессии
model_pairwise = sm.OLS(y, sm.add_constant(X['MSFT']))
results_pairwise = model_pairwise.fit()

# Просмотр результатов оценки линейной модели парной регрессии
print(results_pairwise.summary())
```
