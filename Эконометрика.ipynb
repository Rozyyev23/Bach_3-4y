```python
import numpy as np

# Создаем матрицу коэффициентов
A = np.array([
    [1, 5, 1300, 0, 0, 0],
    [0, 60, 0, 60, 0, 0],
    [1, 4, 1300, 0, 0, 0],
    [0, 0, 0, 62, 0, 63],
    [1, 5, 1600, 0, 0, 0],
    [0, 0, 0, 0, 66, 0]
])

# Создаем вектор значений
B = np.array([60, 5, 62, 4.2, 62, 3.8])

# Решаем систему уравнений
X = np.linalg.solve(A, B)

# Выводим результаты
print("a_01 =", X[0])
print("a_02 =", X[1])
print("a_1 =", X[2])
print("a_2 =", X[3])
print("b_12 =", X[4])
print("b_21 =", X[5])
```
